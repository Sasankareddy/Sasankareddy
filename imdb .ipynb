```python
def to_uppercase(column):
    return column.upper()
```

```python
df['capitalized title']=df['title'].apply(to_uppercase)
df.head(2)
```

```python
def revised_profit(revenue,budget):
    if(revenue-budget)>0:
        return"profit"
    else:
        return "loss"
```

```python
df['profit/loss']=df.apply(lambda x:revised_profit(x['revenue'],x['budget']),axis=1)
df.head(2)
```

```python
df['status_upper']=df['status'].str.upper()
df.head(2)
```

```python
df['title_lower']=df['title'].str.lower()
df.head(2)
```

```python
df['profit/loss_capitalized']=df['profit/loss'].str.capitalize()
df.head(2)
```

```python
df['release_date']=pd.to_datetime(df['release_date'])
df.dtypes
```

```python
df['year']=df['release_date'].dt.year
df.head(2)
```

```python
df['month']=df['release_date'].dt.month
df.head(2)
```

```python
#create a new column"old or new" such that if the year is greater than 2015 print new film else print old film(using apply method)
def year_status(year):
    if(year)>2015:
        return"new film"
    else:
        return "old film"
```

```python
df['new film/old film']=df.apply(lambda x:year_status(x['year']),axis=1)
df.head(2)
```

```python
# find how many movies in the dataset is a flop in the theatre
flop_movies=df[df['profit/loss']=='loss']
flop_movies.shape[0]
```

```python
# find out the details of movies that are not released in the theatre
not_released=df[df['status_upper']=='RUMORED']
not_released.shape[0]
```
