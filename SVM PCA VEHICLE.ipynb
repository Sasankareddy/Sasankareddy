```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_cols
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols
```

```python
df_before = df.copy()
```

```python
for col in categorical_cols:df[col].fillna(df[col].mode()[0], inplace=True)

for col in numeric_cols:df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
import category_encoders as ce
encoder=ce.OrdinalEncoder(cols=categorical_cols)
df=encoder.fit_transform(df)
```

```python
df.dtypes
```

```python
# split the dataset into features and target column
x=df.drop("class",axis=1)
y=df["class"]
```

```python
df["class"].value_counts()
```

```python
x.head()
```

```python
# scaling the input feature data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
x_scaled.head()
```

```python
y.head()
```

```python
# divide into train dataset and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
# applying SVM
from sklearn.svm import SVC
svm_model1 = SVC(kernel='rbf', C=1000.0, gamma='scale', random_state=42)
svm_model1.fit(x_train,y_train)
```

```python
pred1= svm_model1.predict(x_test)
```

```python
#model evaluation
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
rbf_accuracy = accuracy_score(y_test,svm_model1.predict(x_test)) *100

print(f"rbf Accuracy: {rbf_accuracy:.2f}%")
```

```python
print("\nClassification Report:\n", classification_report(y_test,pred1))
print("\nconfusion_matrix:\n", confusion_matrix(y_test,pred1))
```

```python
ce=confusion_matrix(y_test,pred1)
ce
sns.heatmap(ce,annot=True)
```
