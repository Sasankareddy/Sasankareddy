```python
# split the dataset into features and target column
df.columns = df.columns.str.strip()
x=df.drop("Class",axis=1)
y=df["Class"]
```

```python
df["Class"].value_counts()
```

```python
# divide into train dataset and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
#aplly the ML algorithm
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()
lr_model.fit(x_train,y_train)
```

```python
#model prediction
lr_pred = lr_model.predict(x_test)
lr_pred
```

```python
lr_pred[:5]
```

```python
y_test[:5]
```

```python
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
lr_accuracy = accuracy_score(y_test, lr_model.predict(x_test)) * 100
print(f"LR Accuracy: {lr_accuracy:.2f}%")
```

```python
print("\nClassification Report:\n", classification_report(y_test, lr_pred))
```

```python
cm = confusion_matrix(y_test, lr_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Logistic Regression with PCA)")
plt.show()
```

```python
# apply PCA
from sklearn.decomposition import PCA
pca_1 = PCA()
x_train_pca = pca_1.fit_transform(x_train_scaled)

print("Explained Variance Ratio:", pca_1.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8,6))
components = np.arange(1, len(pca_1.explained_variance_ratio_) + 1)
plt.bar(components, pca_1.explained_variance_ratio_ * 100, color='teal')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance (%)')
plt.title('Variance Explained by Each Principal Component')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
pca_2=PCA(n_components=2)
x_test_pca = pca_2.fit_transform(x_test_scaled)
```
