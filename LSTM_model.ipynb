```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM
```

```python
x=[]
y=[]

#sequence1=np.arange(10,110,10)
sequence=[]
for i in range(10,110,10):
    sequence.append(i)

sequence
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
sequence=scaler.fit_transform(np.array(sequence).reshape(-1,1))
sequence
```

```python
#list(sequence1)

for i in range(len(sequence)-3):
    x.append(sequence[i:i+3]) # 3 time steps
    y.append(sequence[i+3])#next value to predict
```

```python
x
```

```python
y
```

```python
x=np.array(x)
y=np.array(y)
```

```python
x.shape
```

```python
x=x.reshape((x.shape[0],x.shape[1],1))
```

```python
x.shape
```

```python
x
```

```python
#lstm model creation
model=Sequential([
    LSTM(50,activation='tanh',input_shape=(3,1)),
    Dense(1)
])
```

```python
# Compile the model
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x,y,epochs=500,verbose=0)
```

```python
# model prediction ex-1
test_input = np.array([70, 80, 90]).reshape(3,1)
test_input=scaler.transform(test_input)
test_input = test_input.reshape((1, 3, 1))

y_pred = model.predict(test_input)
y_pred=scaler.inverse_transform(y_pred)
print(f"predict next is: {y_pred[0][0]:.2f}")
```
