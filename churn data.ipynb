```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_cols
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols
```

```python
df_before = df.copy()
```

```python
for col in categorical_cols:df[col].fillna(df[col].mode()[0], inplace=True)

for col in numeric_cols:df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
import category_encoders as ce
encoder=ce.OrdinalEncoder(cols=categorical_cols)
df=encoder.fit_transform(df)
```

```python
# split the dataset into features and target column
x=df.drop("Churn",axis=1)
y=df["Churn"]
```

```python
df["Churn"].value_counts()
```

```python
x.head()
```

```python
# scaling the input feature data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
X_scaled = pd.DataFrame(X_scaled, columns=x.columns)
X_scaled.head()
```

```python
# divide into train dataset and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
# apply ml algorithm into training data
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#model prediction
y_pred=model.predict(x_test)
```

```python
y_pred[:5]
```

```python
y_test[:5]
```

```python
model.predict_proba(x_test)
```

```python
#model evaluation
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
plt.figure(figsize=(5,2))
sns.heatmap(cm,annot=True,fmt="d")
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,y_pred)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
