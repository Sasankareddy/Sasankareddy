```python
#to check the null values in every columns particularly
for col in df.columns:
    print(f"{col}:{df[col].isnull().sum()}")
```

```python
# Calculate the percentage of null values for each column
null_percentage = (df.isnull().sum() / len(df)) * 100
null_percentage.sort_values(ascending=False)
```

```python
#to check how many null values are there that are less than 60%
for col in df.columns:
    if df[col].isnull().sum()>0.6*len(df):
        print(f"{col}:{df[col].isnull().sum()}")
```

```python
for col in df.columns:
    if df[col].isnull().sum()>0.6*len(df):
        df.drop(col,axis=1,inplace=True)

df.shape
```

```python
for col in df.columns:
    print(f"{col}:{df[col].nunique()}")
```

```python
df.drop(columns=["Order","PID"],axis=1,inplace=True)
df.shape
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_cols
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols
```

```python
df_before = df.copy()
```

```python
for col in categorical_cols:df[col].fillna(df[col].mode()[0], inplace=True)

for col in numeric_cols:df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
#Encoding
!pip install category_encoders
```

```python
import category_encoders as ce
encoder=ce.OrdinalEncoder(cols=categorical_cols)
df=encoder.fit_transform(df)
```

```python
df.shape
```

```python
df.head()
```

```python
# split the dataset into features and target column
x=df.drop("SalePrice",axis=1)
y=df["SalePrice"]
```

```python
x.head()
```

```python
# scaling the input feature data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
X_scaled = pd.DataFrame(X_scaled, columns=x.columns)
X_scaled.head()
```

```python
# divide into train dataset and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
# apply ml algorithm into training data
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#model prediction
y_pred=model.predict(x_test)
```

```python
y_pred[:5]
```

```python
y_test[:5]
```

```python
#Model evaluation
#find r2_score
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)*100
print("r2_score of the model is",round(r2_score(y_test,y_pred)*100,2),"%")
```

```python
# find MAE
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,y_pred)
```

```python
#find MSE
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```

```python
#find RMSE
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(y_test,y_pred)
```
