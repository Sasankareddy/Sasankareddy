SPLIT THE DATA INTO TESTING AND TRAINING

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

APPLY MODEL

```python
from sklearn.svm import SVC
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_model.fit(x_train,y_train)
```

RUN SVM WITH LINEAR KERNAL

```python
svm_model3=SVC(kernel='linear',C=1.0)
svm_model3.fit(x_train,y_train)
```

RUN SVM WITH LINEAR KERNAL AND C=100.0

```python
svm_model2=SVC(kernel='linear',C=100.0)
svm_model2.fit(x_train,y_train)
```

RUN SVM WITH LINEAR KERNAL AND C=1000.0

```python
svm_model4=SVC(kernel='linear',C=1000.0)
svm_model4.fit(x_train,y_train)
```

MODEL PREDICTION

```python
pred1= svm_model.predict(x_test)
pred2= svm_model2.predict(x_test)
pred3= svm_model3.predict(x_test)
pred4= svm_model4.predict(x_test)
```

EVALUATE ACCURACY

```python
#model evaluation
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
rbf_accuracy = accuracy_score(y_test,svm_model.predict(x_test)) *100
lr_accuracy = accuracy_score(y_test,svm_model3.predict(x_test)) *100
lrr_accuracy= accuracy_score(y_test, svm_model2.predict(x_test)) * 100
lrrr_accuracy = accuracy_score(y_test, svm_model4.predict(x_test)) * 100

print(f"rbf Accuracy: {rbf_accuracy:.2f}%")
print(f"lr Accuracy: {lr_accuracy:.2f}%")
print(f"lrr Accuracy: {lrr_accuracy:.2f}%")
print(f"lrrr Accuracy: {lrrr_accuracy:.2f}%")
```

```python
print("\nClassification Report:\n", classification_report(y_test,pred1))
print("\nconfusion_matrix:\n", confusion_matrix(y_test,pred1))
```

#summary::
Project Summary – Pulsar Star Detection using SVM

The project focuses on classifying pulsar stars, which are a rare type of neutron star that emit detectable radio signals on Earth. The goal was to build a predictive model using Support Vector Machines (SVM) to distinguish pulsar stars from non-pulsars based on the HTRU_2 dataset.

The dataset includes 8 numerical features derived from radio wave profiles, such as:

Mean, standard deviation, skewness, and kurtosis of the integrated profile, and

Mean, standard deviation, skewness, and kurtosis of the DM-SNR curve,
along with a binary Class label (1 = Pulsar, 0 = Non-pulsar).

Before training, the data was cleaned, scaled, and split into training (80%) and testing (20%) sets.
The StandardScaler was used to normalize features, ensuring all attributes contribute equally to the model.

Multiple SVM models were trained with different configurations:

RBF kernel (C=1.0)

Linear kernel (C=1.0)

Linear kernel (C=100.0)

Linear kernel (C=1000.0)

Each model’s performance was evaluated using accuracy scores on the test data.

Model Performance Insights

Both Linear kernel with C=1.0 and Linear kernel with C=1000.0 achieved the highest accuracy, indicating that the dataset is linearly separable and does not require complex nonlinear transformations for optimal classification.

The RBF kernel, despite being flexible, did not outperform the linear models — suggesting that the linear decision boundary is sufficient for pulsar classification.

Increasing the regularization parameter C beyond 1000 didn’t provide any accuracy improvement, showing that the model had already achieved its optimal margin.

Conclusion
The analysis shows that SVM with a Linear kernel is the most effective model for this dataset, balancing simplicity and performance. The high accuracy achieved implies that the feature set extracted from pulsar signals is well-structured and linearly discriminative.
This model can be reliably used to identify potential pulsar candidates in large-scale radio signal data with strong predictive confidence.
