```python
df.drop(columns=['Unnamed: 0'],inplace=True)
df.shape
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['International Reputation'].fillna(df['International Reputation'].median(), inplace=True)
```

```python
df['Weak Foot'].fillna(df['Weak Foot'].mode()[0], inplace=True)
```

```python
print(df[['Age', 'International Reputation', 'Weak Foot']].isnull().sum())
```

```python
'''1.  What is the distribution of preferred foot ?
2. What is the distribution of weak foot ?
3. How can we visualize the age distribution of players.
4. How can we track (top 10) international reputation of players on the basis of name.
5. What is the international reputation of left and right footed players, print it with their count?'''
```

```python
#1
preferred_foot_counts = df['Preferred Foot'].value_counts()
preferred_foot_counts
```

```python
preferred_foot_counts.plot(kind='bar', color=['skyblue', 'salmon'])
```

```python
#2
weak_foot_counts = df['Weak Foot'].value_counts().sort_index()
weak_foot_counts
```

```python
weak_foot_counts.plot(kind='bar', color='lightgreen')
```

```python
#3
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#4
top_10_reputation = df.sort_values(by='International Reputation', ascending=False).head(10)
top_10_reputation[['Name', 'International Reputation']]
```

```python
#5
preferred_foot_analysis = df.groupby('Preferred Foot')['International Reputation'].agg(['mean', 'count'])
preferred_foot_analysis
```

```python
'''6. What is the different work rate of the players?
7. How can we visualze work rate of players segregated by preferred foot?
8.  What is the different skill moves of the players?
9. How can we visualze skill moves of players segregated by preferred foot?
10. How can we visualize special score of the players?'''
#6
work_rate_counts = df['Work Rate'].value_counts()
work_rate_counts
```

```python
work_rate_counts.plot(kind='bar', color='purple')
```

```python
#7
work_rate_by_foot = df.groupby(['Preferred Foot', 'Work Rate']).size().unstack(fill_value=0)
work_rate_by_foot
```

```python
work_rate_by_foot.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Work Rate Distribution by Preferred Foot')
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.show()
```

```python
#8
skill_moves_counts = df['Skill Moves'].value_counts().sort_index()
skill_moves_counts
```

```python
skill_moves_counts.plot(kind='bar', color='orange')
plt.title('Distribution of Skill Moves')
plt.xlabel('Skill Moves Rating')
plt.ylabel('Number of Players')
```

```python
#9
skill_moves_by_foot = df.groupby(['Preferred Foot', 'Skill Moves']).size().unstack(fill_value=0)
skill_moves_by_foot 
```

```python
skill_moves_by_foot.plot(kind='bar', stacked=True, figsize=(12, 7))
```

```python
#10
plt.figure(figsize=(10, 6))
plt.hist(df['Special'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Special Scores')
plt.xlabel('Special Score')
plt.ylabel('Number of Players')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#15
df_top_countries = df[df['Nationality'].isin(top_10_countries.index)]
df_top_countries
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(x='Nationality', y='Age', data=df_top_countries)
plt.title('Age Distribution by Top 10 Countries')
plt.xlabel('Nationality')
plt.ylabel('Age')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
