```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN
```

```python
x=[]
y=[]

#sequence1=np.arange(10,110,10)
sequence=[]
for i in range(10,110,10):
    sequence.append(i)

sequence
```

```python
#list(sequence1)

for i in range(len(sequence)-3):
    x.append(sequence[i:i+3]) # 3 time steps
    y.append(sequence[i+3])#next value to predict
```

```python
x
```

```python
y
```

```python
x=np.array(x)
y=np.array(y)
```

```python
x
```

```python
x.shape
```

```python
x=x.reshape((x.shape[0],x.shape[1],1))
```

```python
x
```

```python
# model creation
model=Sequential([
    SimpleRNN(50,activation='relu',input_shape=(3,1)),
    Dense(1)
])
```

```python
# Compile the model
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x,y,epochs=300,verbose=1)
```

```python
# model prediction ex-1
test_input = np.array([80, 90, 100])
test_input = test_input.reshape((1, 3, 1))

prediction = model.predict(test_input)
print(f"predict next is: {prediction[0][0]:.2f}")
```

```python
# ex-2
test_input = np.array([[50, 60, 70]])
test_input = test_input.reshape((1, 3, 1))

prediction = model.predict(test_input)
prediction
```
