```python
#19. How do average Value For Money ratings vary by month in a cyclical (seasonal) pattern?
df['Date Flown'] = pd.to_datetime(df['Date Flown'], format='%b-%y', errors='coerce')
df_clean = df.dropna(subset=['Date Flown', 'Value For Money'])
df_clean['Month'] = df_clean['Date Flown'].dt.strftime('%B')
avg_value_month = df_clean.groupby('Month')['Value For Money'].mean().reset_index()
avg_value_month['Month'] = pd.Categorical(avg_value_month['Month'], 
                                         categories=['January', 'February', 'March', 'April', 'May', 'June', 
                                                     'July', 'August', 'September', 'October', 'November', 'December'], 
                                         ordered=True)
avg_value_month = avg_value_month.sort_values('Month')
avg_value_month
```

Objective:
To calculate the average Value For Money rating for each month to observe any seasonal trends.

Insight:
Reveals cyclical patterns in passenger perceptions of value, showing which months passengers feel they get better or worse value for money

```python
labels = avg_value_month['Month']
values = list(avg_value_month['Value For Money']) + [avg_value_month['Value For Money'].iloc[0]]
angles = [i / 12 * 2 * np.pi for i in range(13)]
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})
ax.plot(angles, values, 'o-', linewidth=2, color='teal')
ax.fill(angles, values, color='teal', alpha=0.3)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
ax.set_title('Average Value For Money by Month', fontsize=14, pad=20)
ax.set_ylim(0, 5)
ax.set_yticks([1, 2, 3, 4, 5]);
```

Objective:
To visualize the monthly variation of Value For Money ratings in a circular (polar) plot, highlighting seasonal trends.

Insight:
Shows cyclical patterns in passenger perception of value, making it easy to identify months with higher or lower satisfaction regarding cost-effectiveness
