```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_cols
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols
```

```python
df_before = df.copy()
```

```python
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

for col in numeric_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# 1. Least 10 popular songs
least_popular_songs = df.nsmallest(10, 'popularity')
least_popular_songs[['name', 'artists', 'popularity']]
```

```python
# 2. 10 most popular songs
most_popular_songs = df.nlargest(10, 'popularity')
most_popular_songs[['name', 'artists', 'popularity']]
```

```python
# 3. Check the artist at 25th row
artist_at_25 = df.loc[24, 'artists']
artist_at_25
```

```python
# 4. Change the duration in milliseconds to seconds
df['duration_sec'] = df['duration_ms'] / 1000
df[['name', 'duration_ms', 'duration_sec']].head()
```

```python
# 5. Regression plot between loudness and energy
plt.figure(figsize=(10, 6))
sns.regplot(x='loudness', y='energy', data=df)
plt.title('Regression Plot: Loudness vs Energy')
plt.show()
```

```python
# 6. Regression plot between Acousticness and popularity
plt.figure(figsize=(10, 6))
sns.regplot(x='acousticness', y='popularity', data=df)
plt.title('Regression Plot: Acousticness vs Popularity')
plt.show()
```

```python
# 7. Distribution plot to visualise total number of songs in each year since 1922
df['release_year'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year
songs_per_year = df[df['release_year'] >= 1922]['release_year'].value_counts().sort_index()
songs_per_year
```

```python
songs_per_year.plot(kind='line')
plt.title('Total Number of Songs Released Each Year (Since 1922)')
plt.xlabel('Release Year')
plt.ylabel('Number of Songs')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 8. Plot Years vs Duration in seconds
# Assuming 'duration_sec' column is already created in a previous step
avg_duration_per_year = df.groupby('release_year')['duration_sec'].mean().sort_index()
```

```python
plt.figure(figsize=(15, 7))
avg_duration_per_year.plot(kind='line')
plt.title('Average Song Duration Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Average Duration (seconds)')
plt.grid(True)
plt.show()
```
