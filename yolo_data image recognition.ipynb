```python
from ultralytics import YOLO
```

```python
model=YOLO('yolov8n.pt')
model.train(data='coco128.yaml',epochs=5,imgsz=640)
```

```python
results=model.predict(source="dog-collage.avif",save='True',conf=0.25)
results[0].show()
```
