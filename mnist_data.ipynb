```python
import tensorflow as tf
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.mnist.load_data()
```

```python
x_test.shape,y_test.shape
```

```python
#scaling the image
x_train=x_train/255
x_test=x_test/255
```

```python
x_train[0]
```

```python
#flattening the scaled image into 1D array
x_train_flattened=x_train.reshape(len(x_train),28*28)
x_test_flattened=x_test.reshape(len(x_test),28*28)
```

```python
x_train_flattened.shape
```

```python
x_test_flattened.shape
```

```python
#model creation using Fnn
model=tf.keras.Sequential([
    tf.keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')])
```

```python
# Compile the model
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# Train the model
model.fit(x_train_flattened,y_train,epochs=10)
```

```python
model.evaluate(x_test_flattened,y_test)
```

```python
#model prediction
y_pred=model.predict(x_test_flattened)
y_pred[0]
```

```python
y_test[0]
```

```python
import numpy as np
np.argmax(y_pred[0])
```

```python
y_test[0]
```

```python
y_predictions=[]
for pred in y_pred:
    y_predictions.append(np.argmax(pred))

y_predictions[:5]
```

```python
#confusion matrix
cm=tf.math.confusion_matrix(y_test,y_predictions)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
#using flatten layer so that we dont have to call.reshape on input dataset
model4=tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(10,activation='sigmoid')])
```

```python
# Compile the model
model4.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# Train the model
model4.fit(x_train,y_train,epochs=5)
```

```python
model4.evaluate(x_test,y_test)
```

```python
#model prediction
y_pred=model4.predict(x_test)
y_pred[0]
```

```python
y_predictions=[]
for pred in y_pred:
    y_predictions.append(np.argmax(pred))

y_predictions[:5]
```

```python
#confusion matrix
cm=tf.math.confusion_matrix(y_test,y_predictions)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d',cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('Confusion Matrix')
plt.show()
```
