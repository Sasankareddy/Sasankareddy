```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Melbourne_housing_FULL - Melbourne_housing_FULL.csv")
```

```python
# Feature Engineering
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 
               'Propertycount', 'Distance', 'CouncilArea', 'Bedroom2', 
               'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']

df = df[cols_to_use]
```

```python
df.dtypes
```

```python
# Handling Missing Values
df.isna().sum()
```

```python
#handling missing values
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)

df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
#encoding
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
# split the dataset into features and target column
x=df.drop("Price",axis=1)
y=df["Price"]
```

```python
df["Price"].value_counts()
```

```python
# divide into train dataset and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
#apply ML model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.score(x_train,y_train)
```

```python
# using ridge regression
from sklearn.linear_model import Ridge
ridge_reg=Ridge()
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```

```python
#using lasso regression
from sklearn.linear_model import Lasso
Lasso_reg=Lasso()
Lasso_reg.fit(x_train,y_train)
```

```python
Lasso_reg.score(x_train,y_train)
```

```python
Lasso_reg.score(x_test,y_test)
```

```python
#tuning the parameters using grid search cv
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

# Initialize the model
lasso2 = Lasso(random_state=2)

# Define the parameter grid
param_grid = {
    'alpha': [100,200],
    'max_iter': [50,100]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=lasso2, 
                           param_grid=param_grid, 
                           scoring='r2', 
                           cv=5, 
                           n_jobs=-1, 
                           verbose=1)
```

```python
# Fit the grid search to your training data
grid_search.fit(x_train, y_train)
```

```python
# Get the best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best R² Score:", grid_search.best_score_)
```

```python
Lasso_reg2=Lasso(alpha=100,max_iter=100)
Lasso_reg2.fit(x_train,y_train)
```

```python
Lasso_reg2.score(x_train,y_train)
```

```python
Lasso_reg2.score(x_test,y_test)
```

#summary::
Overview:
The notebook performs Lasso (L1) and Ridge (L2) regression on the Melbourne Housing Dataset to predict house prices. It explores feature selection, missing value handling, and hyperparameter tuning for better model accuracy.

Key Steps:
Data Loading & Cleaning
The dataset Melbourne_housing_FULL.csv is imported using pandas.
Missing values are handled using fillna() (mean for numeric features, zero for others).
Unnecessary columns are dropped, and the data is filtered for relevant features.

Feature Engineering
Columns such as Rooms, Distance, Landsize, BuildingArea, and Price are selected.
Categorical columns are encoded using one-hot encoding.

Model Implementation
Both Lasso (L1) and Ridge (L2) regressions are applied using sklearn.linear_model.
The dataset is split into training and testing sets using train_test_split.

Parameter Tuning
Hyperparameters like alpha are tuned using GridSearchCV to find the optimal regularization strength.
The best model parameters and R² scores are printed.

Model Evaluation
Metrics used: R² Score, MAE, MSE, RMSE.
Comparison between Lasso and Ridge models shows the effect of regularization on overfitting.

Results Summary:
Lasso Regression reduces less significant feature weights to zero (feature selection).
Ridge Regression performs better when all features are relevant.
The tuned models achieve improved accuracy and generalization.
