```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.fashion_mnist.load_data()
```

```python
#scaling the image
x_train=x_train/255
x_test=x_test/255
```

```python
x_train[0]
```

```python
#flattening the scaled image into 1D array
x_train_flattened=x_train.reshape(len(x_train),28*28)
x_test_flattened=x_test.reshape(len(x_test),28*28)
```

```python
x_train_flattened.shape,x_test_flattened.shape
```

```python
#model creation using Fnn
model1=tf.keras.Sequential([
    tf.keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(10,activation='sigmoid')])
```

```python
# Compile the model
model1.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# Train the model
model1.fit(x_train_flattened,y_train,epochs=10)
```

```python
model1.evaluate(x_test_flattened,y_test)
```

```python
#model prediction
y_pred=model1.predict(x_test_flattened)
y_pred[0]
```

```python
y_test[0]
```

```python
import numpy as np
np.argmax(y_pred[0])
```

```python
y_test[0]
```

```python
y_predictions=[]
for pred in y_pred:
    y_predictions.append(np.argmax(pred))

y_predictions[:5]
```

```python
#confusion matrix
cm=tf.math.confusion_matrix(y_test,y_predictions)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d',cmap='Greens')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('Confusion Matrix')
plt.show()
```
